Aplicar el algoritmo Apriori o una técnica equivalente.

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
datos = {
    'transaccion_id': ['T1','T1','T1','T2','T2','T3','T3'],
    'item': ['Calculadora','novela','Pinceles','Regla','Lapiz','Cartulina','Folder']
}
```

```python
df = pd.DataFrame(datos)

print("DATOS ORIGINALES")
print(df)
```

```python
columnas_asociacion = [
    "transaccion_id",
    "cliente_id",
    "fecha",
    "categoria",
    "item",
    "cantidad",
    "Canal"
]
```

```python
basket = pd.crosstab(
    df['transaccion_id'],
    df['item']
)

print(basket.head())
```

```python
frecuentes = apriori(
    basket,
    min_support=0.05,
    use_colnames=True
)

print("\nITEMSETS FRECUENTES")
print(frecuentes)
```

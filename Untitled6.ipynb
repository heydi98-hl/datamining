```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Prmeramente cargamos el archivo csv asimismo nos da las primeras filas

```python

url="https://raw.githubusercontent.com/heydi98-hl/datamining/refs/heads/main/Archivo/clave_E_correlacion.csv"
df = pd.read_csv(url)
df.head()
```

```python

df.info()
```

```python
df.describe()
```

```python

df.isnull().sum()
```

```python
df_numeric = df.select_dtypes(include=['number'])
```

```python

matriz = df_numeric.corr()
matriz
```

```python

plt.figure(figsize=(10,6))
sns.heatmap(matriz, annot=True, cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```
